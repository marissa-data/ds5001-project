# Interpretation (4)

Describe something interesting about your corpus that you discovered during the process of completing this assignment.

At a minumum, use 250 words, but you may use more. You may also add images if you'd like.

Through completing this assignment, I found that the novels we chose to represent each gothic era of literature (high gothic, romantic/classical gothic, victorian gothic, and modern gothic) do not deterministically predict novel-specific features. This is to be expected, since the novels were a mixture of long and short forms, were all written by different authors with different writing styles, and all tackled diverse themes within the genre of gothic literature. However, through both PCA and sentiment analysis, I discovered that life-threatening fear was much more prevalent in earlier texts (high and romantic/classical gothic eras). The abundance of historical conflicts/wars in the late 1790s to the early 1800s heavily influenced gothic literature authors to encorporate themes of violence and death (often represented through monsters) into their stories (through both realistic fiction and sci-fi/fantasy/horror novels). In comparison, novels written in the late 19th century and beyond (victorian and modern gothic eras) captured the shift to internalized fears, which was largely driven by the increased understanding of mental and emotional health and less conflicts related to physical violence; often these novels focus on personal relationships/experiences as opposed to the widespread horror expressed in earlier texts. Specifically in the LDA and PCA visualization, you can see how diverse these topics within internalized fear are; authors included in this corpus creatively expressed differing aspects of these societial fears through their unique stories, making novel topics less similar statistically. Additionally, since the GI lexicon did not explicitly define many themes related to this more nuanced type of fear, it appears as if fear is not as prevalent of a theme in these later texts. However, both PCA visualizations provide assurance that victorian and modern gothic texts are semantically related, but it is unclear if this is because of similar themes of internalized fear, or because of confounding variables like writing styles and prevalant vocabulary in the later time periods (or hopefully a mixture of both). Overall, it was a fun project to complete and exploring these famous novels through text analysis techniques offered an increased insight into both intended and unintended thematic choices made by each author, without having to read the novels myself.
