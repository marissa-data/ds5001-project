### Load in packages

```python
import pandas as pd
import numpy as np
import plotly.express as px

from gensim.corpora import Dictionary
from gensim.models import word2vec

from sklearn.manifold import TSNE as tsne
```

```python
import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
import gensim
gensim.__version__
```

### Load in CORPUS and set OHCO

```python
OHCO = ['era', 'book_id','chap_num', 'para_num', 'sent_num', 'token_num']
BAG = OHCO[:4]
BAG_2 = OHCO[:3]
```

```python
CORPUS = pd.read_csv('data/CORPUS.csv').set_index(OHCO).dropna()
```

### Convert to Gensim

```python
docs = CORPUS.groupby(BAG).term_str.apply(list).tolist()
docs_2 = CORPUS.groupby(BAG_2).term_str.apply(list).tolist()
```

```python
for i in range(3):
    print(f"Doc {i}:", docs[i])
```

### Extract a dictionary (vocabulary)

```python
dictionary = Dictionary(docs)
dictionary_2 = Dictionary(docs_2)
```

### Generate word embeddings

```python
w2v_params = dict(
    window = 2,
    vector_size = 200,
    min_count = 50,
    workers = 4
)
```

```python
model = word2vec.Word2Vec(docs, **w2v_params)
model.wv.vectors
```

```python
model_2 = word2vec.Word2Vec(docs_2, **w2v_params)
model_2.wv.vectors
```

### Convert model to dataframe

```python
WV = pd.DataFrame(model.wv.vectors, index=model.wv.index_to_key)
WV.index.name = 'term_str'
WV
```

```python
WV_2 = pd.DataFrame(model_2.wv.vectors, index=model_2.wv.index_to_key)
WV_2.index.name = 'term_str'
WV_2
```

### tSNE

```python
PP = 40

tsne_engine = tsne(
    perplexity=PP, 
    n_components=2, 
    init='pca', 
    max_iter=2500, 
    random_state=23
)
TSNE = pd.DataFrame(
    tsne_engine.fit_transform(WV), 
    columns=['x','y'], 
    index=WV.index)
TSNE
```

```python
tsne_engine = tsne(
    perplexity=PP, 
    n_components=2, 
    init='pca', 
    max_iter=2500, 
    random_state=23
)
TSNE_2 = pd.DataFrame(
    tsne_engine.fit_transform(WV_2), 
    columns=['x','y'], 
    index=WV_2.index)
TSNE_2
```

```python
px.scatter(TSNE.reset_index(), 'x', 'y', 
        text='term_str', 
        hover_name='term_str',  
        height=1000,
        width=1200)\
    .update_traces(
        mode='markers+text', 
        textfont=dict(color='black', size=14, family='Arial'),
        textposition='top center')
```

```python
px.scatter(TSNE_2.reset_index(), 'x', 'y', 
        text='term_str', 
        hover_name='term_str',  
        height=1000,
        width=1200)\
    .update_traces(
        mode='markers+text', 
        textfont=dict(color='black', size=14, family='Arial'),
        textposition='top center')
```
